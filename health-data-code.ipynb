```python
# Function to split a single entry
def split_entry(entry):
    # Split by "days."
    primary_split = entry.split(' days.')
    
    # Add "days." back to all elements except the last
    primary_split = [segment + ' days.' if i < len(primary_split) - 1 else segment.strip() for i, segment in enumerate(primary_split)]
    
    # Further split by "," for each primary segment
    split_data = []
    for segment in primary_split:
        if segment:  # Skip empty strings
            sub_segments = segment.split(',')
            split_data.extend([sub.strip() for sub in sub_segments if sub.strip()])
    
    return split_data

def process_csv(input_file, output_file):
    # Read the CSV file
    df = pd.read_csv(input_file)

    # Assuming the column to be processed is named 'data'
    df['medicine'] = df['medicine'].apply(split_entry)

    # Expand the lists into separate columns
    processed_df = df['medicine'].apply(pd.Series)

    # Save the result to a new CSV file
    processed_df.to_csv(output_file, index=False, header=False)

input_file = 'medicine-column.csv'  
output_file = 'split-medicine.csv'  
process_csv(input_file, output_file)


```

```python
def process_and_shift_data(input_file, output_file):
    delimiter = "General Counselling."
    
    chunk_size = 10000  
    chunks = pd.read_csv(input_file, chunksize=chunk_size)
    
    processed_chunks = []

    for df in chunks:
        # Iterate over the DataFrame rows
        for index, row in df.iterrows():
            for i in range(1, 10): 
                drug_col = f'drug_{i}'
                dosage_col = f'dosage_{i}'
                duration_col = f'duration_{i}'
                
                if pd.notna(row[drug_col]) and delimiter in str(row[drug_col]):
                    # Split the data into counselling and medication parts
                    counselling, medication_info = row[drug_col].split(delimiter, 1)
                    
                    # Update the column with the counselling part
                    df.at[index, drug_col] = delimiter.strip()
                    
                    # Shift subsequent columns to the right by one set of columns
                    for j in range(9, i, -1):  # Assuming up to 'drug_9', 'dosage_9', 'duration_9'
                        df.at[index, f'drug_{j+1}'] = df.at[index, f'drug_{j}']
                        df.at[index, f'dosage_{j+1}'] = df.at[index, f'dosage_{j}']
                        df.at[index, f'duration_{j+1}'] = df.at[index, f'duration_{j}']
                    
                    # Place the medication info into the next set of columns
                    df.at[index, f'drug_{i+1}'] = medication_info.strip()
                    df.at[index, f'dosage_{i+1}'] = row[dosage_col]
                    df.at[index, f'duration_{i+1}'] = row[duration_col]
                    
                    # Clear the original dosage and duration cells for the counselling
                    df.at[index, dosage_col] = ''
                    df.at[index, duration_col] = ''
                    
                    break  # Stop after finding the first occurrence of "General Counselling."

        # Collect the processed chunk
        processed_chunks.append(df)
    
    # Concatenate all processed chunks into a single DataFrame
    result_df = pd.concat(processed_chunks, ignore_index=True)
    
    # Save the result to a new CSV file
    result_df.to_csv(output_file, index=False)

# Input and output file paths
input_file = 'split-medicine.csv'
output_file = 'split-medicine.csv'

# Call the function to process and shift data
process_and_shift_data(input_file, output_file)




```

```python
import pandas as pd

chunk_size = 100000  
reference_df = pd.read_csv('unique-1.csv')

replacement_dict = dict(zip(reference_df['Drugs'], reference_df['Corrected Values']))

header_written = False

with pd.read_csv('updated_data.csv', chunksize=chunk_size) as reader:
    for chunk in reader:
        # Replace values in the chunk based on the dictionary
        chunk.replace(replacement_dict, inplace=True)
        
        # Append to the cleaned CSV (write header only for the first chunk)
        chunk.to_csv('sample-data.csv', mode='a', index=False, header=not header_written)
        
        # After the first chunk, set header_written to True
        header_written = True

print("Large CSV cleaned and saved as 'cleaned_file.csv'")


```
